```python
# Importar librer铆as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuraci贸n para gr谩ficos en alta resoluci贸n
plt.rcParams['figure.dpi'] = 100

# Cargar el dataset
data = pd.read_excel('data.xlsx')

# Mostrar las primeras filas para verificar la estructura
print(data.head())

#Gr谩fico de Barras: Distribuci贸n de Categor铆as de Comentarios usando la columna 'topic_c'
topic_counts = data['Topic_c'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=topic_counts.index, y=topic_counts.values, palette='viridis')
plt.title('Distribuci贸n de Categor铆as de Comentarios')
plt.xlabel('Categor铆a (topic_c)')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('bar_chart.png')
plt.show()

# Gr谩fico de Densidad: Distribuci贸n de la Longitud de Comentarios
data['Longitud'] = data['Comment'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(8, 6))
sns.kdeplot(data['Longitud'], shade=True, color='blue')
plt.title('Distribuci贸n de Longitud de Comentarios')
plt.xlabel('N煤mero de Palabras')
plt.ylabel('Densidad')
plt.tight_layout()
plt.savefig('density_plot.png')
plt.show()

# Mapa de Calor: Correlaci贸n entre Variables Num茅ricas
numeric_cols = data.select_dtypes(include=[np.number])
corr_matrix = numeric_cols.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Variables Num茅ricas')
plt.tight_layout()
plt.savefig('heatmap.png')
plt.show()
```
